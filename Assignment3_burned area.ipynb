##Mission5: 단일변수 선형 회귀분석을 통하여 가장 영향이 높은 속성 선발

```python
sdfreg = df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']]
sdfreg
```

```python
feature_names = ('FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain')
predicator_analysis = {}
for i in range(0, 7):
    predicator_analysis[i] = Series(np.empty(6), index=['slope', 'intercept', 'r_value', 'r_squared', 'p_value', 'std_err'])
    predicator_analysis[i][0],\
    predicator_analysis[i][1],\
    predicator_analysis[i][2],\
    predicator_analysis[i][4],\
    predicator_analysis[i][5] = stats.linregress(sdfreg[feature_names[i]].values, sdfreg['area'].values)
    predicator_analysis[i][3] = predicator_analysis[i][2] ** 2
    
format1 = "%5s %10s %10s %10s %10s %10s %10s"
format2 = "%5s %10f %10f %10f %10f %10f %10f"
print format1 % ('', 'slope', 'intercept', 'r_value', 'r_squared', 'p_value', 'std_err')
for i in range(0, 7):
    lst = [feature_names[i]]
    for j in range(6):
        lst.append(predicator_analysis[i][j])
    print format2 % tuple(lst)
```

##Mission6. 가장 영향이 높은 요인 두 개로 다변수 선형 회귀분석 수행

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
X = zip(sdfreg['temp'], sdfreg['RH'])
y = sdfreg['area'].values
regr = regr.fit(X, y)

print 'Coefficients:', regr.coef_
print ' Intercept:', regr.intercept_
```

선형회귀식: y = 1.9288 + 0.8811 * temp - 0.1293 * RH

```python
# predicted area 값을 한꺼번에 출력하여 비교
predicted = []
for i in range(0, len(sdfreg)):
    predicted.append(regr.predict(X[i]))
    
sdfreg.loc[:, 'predicted'] = predicted
sdfreg
```

선형 회귀분석 성능 좋은지 판단

##Mission7.가장 영향이 높은 요인 두 개로 다변수 로지스틱 회귀분석 수행

```python
sdflogreg = df[['temp', 'RH', 'area']]
sdflogreg.loc[:, 'area'] = np.log1p(sdflogreg['area'])

# 50% 이상이면 1, 아닌 경우 0으로 set
fire = []
for i in range(0, len(sdflogreg)):
    if sdflogreg.iloc[i]['area'] >= 0.418710:
        fire.append(1)
    else:
        fire.append(0)
        
sdflogreg.loc[:, 'fire'] = fire
sdflogreg
```

```python
X = zip(sdflogreg['temp'], sdflogreg['RH'])
y = sdflogreg['fire'].values
regr2 = linear_model.LogisticRegression()
regr2 = regr2.fit(X, y)

print 'Coefficients:', regr2.coef_
print ' Intercept:', regr2.intercept_
```

```python
predicted2 = []
for i in range(0, len(sdflogreg)):
    predicted2.append(regr2.predict(X[i]))
    
sdflogreg.loc[:, 'predicted2'] = predicted2
sdflogreg
```

```python
# 로지스틱 회귀분석에 의한 분류 정확도
correctly_predicted = 0
for i in range(0, len(sdflogreg)):
    if predicted2[i] == y[i]:
        correctly_predicted += 1
        
print float(correctly_predicted) / len(sdflogreg)
```
