```python
import urllib2
from scipy import stats
from pandas import Series, DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/forest-fires/forestfires.csv'
raw_csv = urllib2.urlopen(path)
feature_names = ('X', 'Y', 'month', 'day', "FFMC", "DMC", 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain')
target_name = 'area'
all_names = feature_names + (target_name,)
df = pd.read_csv(raw_csv)
```

```python
df
```

```python
sdf = df[['X', 'Y', 'area']]
sdf
```

## Mission1: 불이 가장 많이 발생한 지역 좌표 (X, Y) 상위 5개

좌표별 area 합 - groupby로 좌표 그룹화, sum()으로 각 좌표의 area 합 구함      

```python
groupby_sdf = sdf.groupby(['X', 'Y']).sum()
groupby_sdf
```

area 합을 정렬 - sort() 이용

```python
sort_sdf = groupby_sdf.sort(target_name, ascending=False)
sort_sdf
```

상위 5개의 좌표 추출

```python
stacked_sdf = sort_sdf.stack()
stacked_sdf
```

```python
stacked_sdf.pop
```

```python
top5_sdf = stacked_sdf.head(5)
top5_sdf
```
